```python
import csv
from matplotlib import pyplot as plt
from datetime import datetime
```

# ACELERANDO O CONHECIMENTO

```python
filename = 'vale_da_morte.csv'
```

```python
with open(filename) as f:
    reader = csv.reader(f)
    reader_top = next(reader)
    
    dates, temp_max, temp_min = [], [], []
    
    for x in reader:
        
        try:
            form_date = datetime.strptime(x[0], "%Y-%m-%d")
            temps_max = (float(x[1])/30)
            temps_min = (float(x[3])/30)
            
        except ValueError:
            print(form_date, 'Vazio')
            
        else:
            dates.append(form_date)
            temp_max.append(temps_max)
            temp_min.append(temps_min)
        
fig = plt.figure(dpi=128, figsize=(14,6))
plt.plot(dates, temp_max, c='red')
plt.plot(dates, temp_min, c='blue')
plt.fill_between(dates, temp_max, temp_min, facecolor='green', alpha=0.3)

plt.title("Temperatura X Período", fontsize=30)
plt.xlabel("", fontsize=15)
fig.autofmt_xdate()
plt.ylabel("Temperatura (C)", fontsize=15)
plt.tick_params(axis='both', which='major', labelsize=15)

plt.show()
```
